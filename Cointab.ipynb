**Categorzing and Assiging the Weight Range Slab.**

```python
def weight_slab(weight):
    decimal_part = weight - int(weight)

    if decimal_part == 0.0:
        return weight
    elif decimal_part > 0.5:
        return int(weight) + 1.0
    else:
        return int(weight) + 0.5
```

**Slab Weight Generation**

```python
renameSKU['Weight Slab As Per Company X (KG)']=renameSKU['Total Weight AS Per Company X (KG)'].apply(weight_slab)
CourierInvoice['Weight Slab Charged By Courier Company (KG)']=(CourierInvoice['Total Weight As Per Courier Company (KG)']).apply(weight_slab)
renameSKU.head()
```

**Calculation of Expected Charges Based on Delivery Zone and Weight Slab.**

```python
def calculate_expected_charge(row, courier_rates):
    zone = row['Delivery Zone as Per Company X']
    fwd_category = 'fwd_' + zone
    rto_category = 'rto_' + zone

    if fwd_category + '_fixed' in courier_rates.columns and fwd_category + '_additional' in courier_rates.columns:
        fwd_fixed = courier_rates.at[0, fwd_category + '_fixed']
        fwd_additional = courier_rates.at[0, fwd_category + '_additional']
    else:
        return 0  # Handle missing keys

    if rto_category + '_fixed' in courier_rates.columns and rto_category + '_additional' in courier_rates.columns:
        rto_fixed = courier_rates.at[0, rto_category + '_fixed']
        rto_additional = courier_rates.at[0, rto_category + '_additional']
    else:
        return 0  # Handle missing keys

    weight_slab = row['Weight Slab As Per Company X (KG)']
    shipment_type = row['Type of Shipment']

    if weight_slab <= 0.5:
        if shipment_type == 'Forward charges':
            return fwd_fixed
        elif shipment_type == 'Forward and RTO charges':
            return fwd_fixed + rto_fixed
    else:
        additional_weight = (weight_slab - 0.5) / 0.5
        if shipment_type == 'Forward charges':
            return fwd_fixed + additional_weight * fwd_additional
        elif shipment_type == 'Forward and RTO charges':
            return fwd_fixed + additional_weight * (fwd_additional + rto_additional)

    return 0  # Default case if conditions are not met

# Calculate and assign expected charges
renameSKU['Expected Charge As Per Company X (Rs.)'] = renameSKU.apply(
    lambda row: calculate_expected_charge(row, Rates),
    axis=1)
```

```python
Resultant_Output=Resultant_Output.rename(columns={'AWB Code':'AWB Number','Billing Amount (Rs.)':'Charges_Billed_By_Courier_Company_Rs','Expected Charge As Per Company X (Rs.)':'Expected_Charge_As_Per_Company_Rs'})
Resultant_Output.head()
```

**Calculation of Variance Between Expected Charges and Billed Amounts**





```python
diff_bw_ec_bc="""
Difference_Between_Expected_Charges_and_Billed_Charges_Rs =(Expected_Charge_As_Per_Company_Rs - Charges_Billed_By_Courier_Company_Rs)
"""
Resultant_Output.eval(diff_bw_ec_bc,inplace=True,engine='python')
```

```python
Resultant_Output=Resultant_Output.rename(columns={'Charges_Billed_By_Courier_Company_Rs':'Charges Billed By Courier Company (Rs.)','Expected_Charge_As_Per_Company_Rs':'Expected Charge As Per Company X (Rs.)','Difference_Between_Expected_Charges_and_Billed_Charges_Rs':'Difference Between Expected Charges And Billed Charges (Rs.)'})
```

```python
writer = pd.ExcelWriter('Expected_Result.xlsx',if_sheet_exists="replace",engine='openpyxl',mode="a")
```

```python
Resultant_Output.to_excel(writer,sheet_name='Calculations')
calculations=writer.save()
```

**Calculation of Total Orders**

```python
correctly_charged=Resultant_Output[Resultant_Output['Difference Between Expected Charges And Billed Charges (Rs.)']==0.0]
correctly=["Total No of Accurate Charges",len(correctly_charged),sum(correctly_charged['Charges Billed By Courier Company (Rs.)'])]

over_charged=Resultant_Output[Resultant_Output['Difference Between Expected Charges And Billed Charges (Rs.)']<0.0]
over=["Total No of Excessive Charges ",len(over_charged),np.absolute(sum(over_charged['Difference Between Expected Charges And Billed Charges (Rs.)']))]

under_charged=Resultant_Output[Resultant_Output['Difference Between Expected Charges And Billed Charges (Rs.)']>0.0]
under=["Total No of Under Charges",len(under_charged),np.absolute(sum(under_charged['Difference Between Expected Charges And Billed Charges (Rs.)']))]
```

```python
Resultant_Output2=pd.DataFrame([correctly,over,under],columns=['Description','Count','Amount (Rs.)'])
Resultant_Output2
```

```python
Resultant_Output2.to_excel(writer,sheet_name='Summary')
writer.save()
```
